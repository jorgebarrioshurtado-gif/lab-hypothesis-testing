**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# H0: dragon_type_hp_mean <= grass_type_hp_mean
# H1: dragon_type_hp_mean > grass_type_hp_mean
```

```python
dragon_hp = df[df["Type 1"] == "Dragon"]["HP"]
grass_hp = df[df["Type 1"] == "Grass"]["HP"]
```

```python
st.ttest_ind(dragon_hp, grass_hp, alternative = "greater")

# the p_value is smaller than 0.05, which means it is actually so improbable that was just by chance that there has to be a reason behind that.
# so we can accept our alternative hypothesis. 
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# H0: legen_stats = no_legen_stats
# H1: legen_stats != no_legen_stats
```

```python
legen_stats = df[df["Legendary"] == True][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
no_legen_stats = df[df["Legendary"] == False][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
```

```python
st.ttest_ind(legen_stats, no_legen_stats, alternative = "two-sided")
# we want to know if the stats are different so now the it is a two tails test
# all the p-values are super tiny so we can reject the null hypotehesis
```

In this challenge, we will be working with california-housing data. The data can be found here:

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import math
```

```python
def euclidean_dist (df):
    school_coordinates = (-118, 34)
    hospital_coordinates = (-122, 37)
    euc_dist_from_school = []
    euc_dist_from_hospital = []
    for i in df.index:
        euc_dist_from_school.append(math.dist((df.loc[i]["longitude"], df.loc[i]["latitude"]), school_coordinates))
        euc_dist_from_hospital.append(math.dist((df.loc[i]["longitude"], df.loc[i]["latitude"]), hospital_coordinates))
    df["euc_dist_from_school"] = euc_dist_from_school
    df["euc_dist_from_hospital"] = euc_dist_from_hospital
    return df

df = euclidean_dist(df)
df.head()
```

```python
close_from_home = df[(df["euc_dist_from_school"] < 0.5) | (df["euc_dist_from_hospital"] < 0.5)]
far_from_home = df[(df["euc_dist_from_school"] >= 0.5) | (df["euc_dist_from_hospital"] >= 0.5)]
```

```python
# We posit that houses close to either a school or a hospital are more expensive.
# H0: close_from_home_price <= far_from_home_price
# H1: close_from_home_price > far_from_home_price
```

```python
st.ttest_ind(close_from_home["median_house_value"], far_from_home["median_house_value"], alternative = "greater")

# the p_value is less than 0.05, so we can reject the null hypothesis and accept the alternative hypothesis
```
